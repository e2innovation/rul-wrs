```
def multiply_by_menosuno(arr):
    return np.multiply(arr,-1)
def mul_menosuno_y_div_diez(arr):
    return np.multiply(arr,-0.1)
def conv_digmode(arr):
    return arr & 2
def conv_dipper_trip(arr):
    return arr & 1

var_imp['HOIST_ROPE_LENGTH_INV'] = var_imp[['HOIST_ROPE_LENGTH']].apply(multiply_by_menosuno)
var_imp['HOIST_SPEED_INV'] = var_imp.loc[:,'HOIST_SPEED'].apply(mul_menosuno_y_div_diez)
var_imp['DIG_MODE'] = var_imp.loc[:,'STATUS_BYTE_3'].apply(conv_digmode) # nos indica que se encuentra en modo excavación
var_imp['DIPEER_TRIP'] = var_imp.loc[:,'STATUS_BYTE_3'].apply(conv_dipper_trip) # se activa con la orden de apretura de compuerta de balde
var_imp['HOIST_TORQUE_INV'] = var_imp[['HOIST_TORQUE']].apply(multiply_by_menosuno)
var_imp['HOIST_IW_INV'] = var_imp[['HOIST_IW']].apply(multiply_by_menosuno)
```

```
# transformando al formato time series 
var_imp['time']= var_imp['DATE_TIME'].apply(pd.Timestamp)
var_imp=var_imp.set_index('time')
var_imp.index[8]
```

La función **detect_ciclo_normalizado** toma como entrada un data frame (_df) con data **normalizada** el cuál contenga como 
- **índice** (index) la fecha y hora
- **'HOIST_SPEED_INV'** = la velocidad de Hoist invertida
- **'HOIST_ROPE_LENGTH_INV'** = la longitud del cables de hoist invertida
- **'DIG_MODE'** = el modo dig cuando esta en (excavación = 2), modo (provel o avance = 0)
- **'HOIST_IW_INV'** = la corriente IW de hoist invertida

**Devuelve:**
- el vector **puntos_ciclos** que contiene tuplas con la fecha y hora inicial y final de cada ciclo detectado
- **puntos_inicios**: contiene la fecha y hora inicial de cada ciclo
- **puntos_finales**: contiene la fecha y hora final de cada ciclo
- **puntos_T2_ini**: contiene la fecha y hora del final de la etapa 1 e inicio de la etapa 2
- **puntos_T3_ini**: contiene la fecha y hora del final de la etapa 2 e inicio de la etapa 3


*Para un mejor análisis el ciclo de carguío se ha dividido en tres etapas*

- **Etapa 1**: se inicia cuando el cucharon se encuentra en la parte inferior cerca al suelo y se comienza a cargar,
culmina cuando el cucharon se encuentra cargado y en la parte superior del banco o frente de carguío.

- **Etapa 2**: se inicia cuando el cucharon se encuentra cargado y se retira del frente de carguío para girar hacia la tolva del camión
y culmina cuando se posiciona sobre la tolva del camión y se de la activación de apertura de la compuerta, para vaciar la carga del
cucharon en la tolva del camión.

- **Etapa 3**: se inicia cuando se aperturo la compuerta del cucharon para vaciar le material en la tolva del camión, y el cucharon
comienza a girar hacia el frente de carguío y culmina cuando el mismo se encuentra en la parte inferior para iniciar un nuevo ciclo de carguío

**Condiciones de detección de**:
- Inicio de ciclo de carguío:
la condición para detectar el punto de inicio de cada ciclo es que el la velocidad de hoist invertida pase de un valor negativoa positivo
ademas la posición del cable de hoist inv sea menor a 30 (es decir que el cucharon o balde se encuentre cerca al suelo)
y que el modo de trabajo de la pala este en modo dig (excavación == 2)

- Fin de ciclo de carguío:
Para la condición del punto final de cada ciclo es que se detecte un nuevo punto de inicio de ciclo o se detecte
un cambio en el modo de trabajo de la pala de dig a propel es decir que la variable 'DIG_MODE' cambie de valor 2 a 0

- Fin de etapa 1 e inicio de etapa 2
la detección del fin de la etapa 1 e inicio de la etapa 2 se da cuando ('HOIST_IW_INV' <= 'HOIST_ROPE_LENGTH_INV') and ['HOIST_ROPE_LENGTH_INV'] >= 0.4
pero se tiene que haber detectado antes un punto de inicio de ciclo

- Fin de etapa 2 e inicio de etapa 3
La detección del fin de la etapa 2 e inicio de la etapa 3 se da cuando se detecta la activación del dipper trip ['DIPEER_TRIP'] == 1
pero se tiene que haber detectado antes un punto de de inicio de etapa 2


```
# Función para la detección de ciclos en base a data normalizada

def detect_ciclo_normalizado(_df):
  inicio = True
  puntos_ciclos = []
  puntos_inicios = []
  puntos_finales = []
  puntos_T2_ini = []
  puntos_T3_ini = []
  for index, row in _df.iterrows():
      if inicio == True:
          vel_hoist_anterior = row['HOIST_SPEED_INV']
          dig_mode_anterior = row['DIG_MODE']
          temp =[]
          inicio_fin =[]
          inicio = False
          T2 = False
          T3 = False
      else:
          vel_hoist_actual = row['HOIST_SPEED_INV']
          hoist_rope = row['HOIST_ROPE_LENGTH_INV']
          dig_mode_actual = row['DIG_MODE']
          if (vel_hoist_anterior < 0.502 and vel_hoist_actual >= 0.502 and hoist_rope < 0.4616) and (dig_mode_actual == 1):
              temp.append(index_anterior)
              puntos_inicios.append(index)
              if len(temp) > 1:
                  puntos_ciclos.append(temp)
                  puntos_finales.append(index_anterior)
                  temp = []
                  temp.append(index_anterior)
                  T2 = False
                  T3 = False
          elif len(temp) == 1 and T2 == False:
                  if (row['HOIST_IW_INV'] <= row['HOIST_ROPE_LENGTH_INV']) and row['HOIST_ROPE_LENGTH_INV'] >= 0.4:
                      puntos_T2_ini.append(index)
                      T2 = True
          elif len(temp) == 1 and T2 == True and T3 == False:
                  if row['DIPEER_TRIP'] == 1:
                      puntos_T3_ini.append(index)
                      T3 = True
              
          elif (dig_mode_anterior == 1 and dig_mode_actual == 0): #detecta el cambio de modo dig(excavación) a modo propel (avance)
              temp.append(index_anterior)
              if len(temp) > 1:
                  puntos_ciclos.append(temp)
                  puntos_finales.append(index_anterior)
                  temp = []
                  T2 = False
                  T3 = False
          vel_hoist_anterior = row['HOIST_SPEED_INV']
          dig_mode_anterior = row['DIG_MODE']
          index_anterior = index
  return puntos_ciclos, puntos_inicios, puntos_finales, puntos_T2_ini, puntos_T3_ini
```

```
vector_N,p_ini_N,p_fin_N,P2_N,P3_N = detect_ciclo_normalizado(test_graf)
```
